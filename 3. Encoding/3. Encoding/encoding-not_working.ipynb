```python
def nothotter(x): #emp city
    d = {}
    n = len(x)
    for a in range(0, n):
        d[x[a]] = a
    return d
```

```python
def onehotter(x): #expects unique array
    d  = {}
    n = len(x)
    for a in range(0, n):
        s = '0 '*a + '1 ' + '0 '*(n-a-1)
        s = s.split(" ")
        l=len(s)-1
        s = s[:l]
        print(s,'\n')
        u = [int(float(a)) for a in s]
        print(u,'\n')
        d[x[a]] = u
    return d
```

```python
def encoder(df):
    new_df = df
    
    #certified/denied
    new_df.loc[new_df['case_status'] == 'Certified', 'case_status'] = 1
    new_df.loc[new_df['case_status'] == 'Denied' ,'case_status'] = 0
    
    #country one hot
    d = onehotter(new_df['country_of_citizenship'].unique())
    new_df['country_of_citizenship'] = new_df['country_of_citizenship'].apply(lambda x: d[x])
    
    #employer cities, drop in logistic
    d1 = onehotter(new_df['employer_state'].unique())
    new_df['employer_state'] = new_df['employer_state'].apply(lambda x: d1[x])
    
    #education
    d2 = onehotter(new_df['foreign_worker_info_education'].unique())
    new_df['foreign_worker_info_education'] = new_df['foreign_worker_info_education'].apply(lambda x: d2[x])
    
    d1 = nothotter(new_df['foreign_worker_info_inst'].unique())
    new_df['foreign_worker_info_inst'] = new_df['foreign_worker_info_inst'].apply(lambda x: d1[x])
    
    
    d1 = nothotter(new_df['foreign_worker_info_major'].unique())
    new_df['foreign_worker_info_major'] = new_df['foreign_worker_info_major'].apply(lambda x: d1[x])
    
    d1 = onehotter(new_df['foreign_worker_info_state'].unique())
    new_df['foreign_worker_info_state'] = new_df['foreign_worker_info_state'].apply(lambda x: d1[x])
    
    d1 = onehotter(new_df['pw_soc_title'].unique())
    new_df['pw_soc_title'] = new_df['pw_soc_title'].apply(lambda x: d1[x])
    
    d1 = nothotter(new_df['employer_city'].unique())
    new_df['employer_city'] = new_df['employer_city'].apply(lambda x: d1[x])
    
    d1 = onehotter(new_df['employer_decl_info_title'].unique())
    new_df['employer_decl_info_title'] = new_df['employer_decl_info_title'].apply(lambda x: d1[x])
    return new_df
```

```python
data = pd.read_csv('../Dataset/dweepafinal.csv')
def statemaker(x):
    try:
        y = len(x)
    except:
        y = -1
    if(y<3 and y!=-1):
        return states[x].lower()
    elif y>2:
        return x.lower()
    return x
data['employer_state'] = data['employer_state'].apply(statemaker)
data['foreign_worker_info_state'] = data['foreign_worker_info_state'].apply(statemaker)
new_encoder = encoder(data)
```

```python
agents  = new_encoder['agent_firm_name'].unique()
agents_success = {}

for a in agents:
    print(a)
    try:
        x = Counter(new_encoder['agent_firm_name']==a)
        y = Counter(new_encoder[new_encoder['agent_firm_name']==a]['case_status']==1)
        try:
            c = y[True]
        except:
            c = 0
        agents_success[a] = c/x[True]
        if x[True]<10:
            agents_success[a] = 0.5
    except:
        continue
    
#     d = dict(Counter(x))
#     agent_success[a] = d
```

```python
def get_success(x):
    try:
        y = agents_success[x]
    except:
        y = 0
    return y
new_encoder['agent_success'] = new_encoder['agent_firm_name'].apply(get_success)
```
