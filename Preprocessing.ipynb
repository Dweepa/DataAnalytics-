```python
data = pd.read_csv("short_visa.csv", low_memory=False)
```

```python
data.drop(['employer_address_1', 'employer_address_2', 'Unnamed: 0', 'Unnamed: 0.1', 'full_count'], axis = 1, inplace = True)
```

```python
data['class_of_admission'].unique()
```

```python
data.columns
```

```python
h1b = data[data['class_of_admission']=='H-1B']
f1 = data[data['class_of_admission']=='F-1']
l1 = data[data['class_of_admission']=='L-1']
```

```python
new_data = pd.concat([h1b, f1, l1])
```

```python
new_data
```

```python
new_data.drop(['pw_amount_9089', 'pw_track_num', 
               'pw_job_title_9089', 'preparer_info_title','preparer_info_emp_completed', 'job_info_major', 
               'job_info_education', 'job_info_foreign_ed', 'job_info_job_title'], axis = 1, inplace = True)
```

```python
new_data.columns
```

```python
new_data.loc[new_data['case_status']=='Certified-Expired','case_status'] = 'Certified'
```

```python
def fun(x):
    try:
        y = float(x.replace(',',''))
    except:
        y=0
    return y
    
new_data['wage_offer_from_9089']=new_data['wage_offer_from_9089'].apply(fun)
new_data
#new_data['wage_offer_from_9089'] = new_data['wage_offer_from_9089'].astype(float)
```

```python
new_data.loc[new_data['wage_offer_unit_of_pay_9089']=='Week','wage_offer_from_9089'] = new_data['wage_offer_from_9089']*52
```

```python
new_data.loc[new_data['wage_offer_unit_of_pay_9089']=='Bi-Weekly','wage_offer_from_9089'] = new_data['wage_offer_from_9089']*26
```

```python
new_data.loc[new_data['wage_offer_unit_of_pay_9089']=='Hour','wage_offer_from_9089'] = new_data['wage_offer_from_9089']*52*8
```

```python
new_data.loc[new_data['wage_offer_unit_of_pay_9089']=='Month','wage_offer_from_9089'] = new_data['wage_offer_from_9089']*12
```

```python
new_data.drop(['wage_offer_unit_of_pay_9089'], axis = 1, inplace = True)
new_data
```

```python
new_data['pw_level_9089'].unique()
```

```python
d = {'Level IV':4, 'Level I':1, 'Level II':2, 'Level III':3}
def fun1(x):
    try:
        y = d[x]
    except:
        y = 0
    return y

new_data['pw_level_9089'] = new_data['pw_level_9089'].apply(fun1)
```
